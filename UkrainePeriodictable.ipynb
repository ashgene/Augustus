```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define element data (Atomic Number, Name, Weight, Position, Category)
elements = {
    "Li": (3, "Lithium", 6.94, (2, 1), "Alkali Metal"),
    "Be": (4, "Beryllium", 9.01, (2, 2), "Alkaline Earth Metal"),
    "Sc": (21, "Scandium", 44.96, (4, 3), "Rare Earth"),
    "Ti": (22, "Titanium", 47.87, (4, 4), "Transition Metal"),
    "Mn": (25, "Manganese", 54.94, (4, 7), "Transition Metal"),
    "Fe": (26, "Iron", 55.85, (4, 8), "Transition Metal"),
    "Co": (27, "Cobalt", 58.93, (4, 9), "Transition Metal"),
    "Ni": (28, "Nickel", 58.69, (4, 10), "Transition Metal"),
    "Cu": (29, "Copper", 63.55, (4, 11), "Transition Metal"),
    "Zn": (30, "Zinc", 65.38, (4, 12), "Transition Metal"),
    "Ga": (31, "Gallium", 69.72, (4, 13), "Other Metal"),
    "Y": (39, "Yttrium", 88.91, (5, 3), "Rare Earth"),
    "Ag": (47, "Silver", 107.87, (5, 11), "Transition Metal"),
    "La": (57, "Lanthanum", 138.91, (6, 3), "Rare Earth"),
    "Ce": (58, "Cerium", 140.12, (6, 4), "Rare Earth"),
    "Pr": (59, "Praseodymium", 140.91, (6, 5), "Rare Earth"),
    "Nd": (60, "Neodymium", 144.24, (6, 6), "Rare Earth"),
    "Sm": (62, "Samarium", 150.36, (6, 7), "Rare Earth"),
    "Eu": (63, "Europium", 151.96, (6, 8), "Rare Earth"),
    "Gd": (64, "Gadolinium", 157.25, (6, 9), "Rare Earth"),
    "Tb": (65, "Terbium", 158.93, (6, 10), "Rare Earth"),
    "Dy": (66, "Dysprosium", 162.50, (6, 11), "Rare Earth"),
    "Ho": (67, "Holmium", 164.93, (6, 12), "Rare Earth"),
    "Er": (68, "Erbium", 167.26, (6, 13), "Rare Earth"),
    "Tm": (69, "Thulium", 168.93, (6, 14), "Rare Earth"),
    "Yb": (70, "Ytterbium", 173.05, (6, 15), "Rare Earth"),
    "Lu": (71, "Lutetium", 174.97, (6, 16), "Rare Earth"),
    "U": (92, "Uranium", 238.03, (7, 3), "Actinide"),
    "Al": (13, "Aluminum", 26.98, (3, 13), "Other Metal"),
    "Pb": (82, "Lead", 207.2, (6, 17), "Other Metal"),
    "Ne": (10, "Neon", 20.18, (2, 18), "Noble Gas"),
    "Kr": (36, "Krypton", 83.80, (4, 18), "Noble Gas"),
    "C": (6, "Graphite", 12.01, (2, 14), "Nonmetal"),
}

# Define color mapping for element categories
category_colors = {
    "Alkali Metal": "#FFD700",
    "Alkaline Earth Metal": "#FF8C00",
    "Transition Metal": "#4682B4",
    "Rare Earth": "#32CD32",
    "Actinide": "#8A2BE2",
    "Other Metal": "#A52A2A",
    "Noble Gas": "#FF4500",
    "Nonmetal": "#708090"
}

# Create figure
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xlim(0.5, 18.5)
ax.set_ylim(-8.5, -0.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Add elements as rectangles
for symbol, (atomic_number, name, weight, (row, col), category) in elements.items():
    color = category_colors.get(category, "gray")
    rect = patches.Rectangle((col-0.5, -row-0.5), 1, 1, linewidth=2, edgecolor="black", facecolor=color)
    ax.add_patch(rect)
    ax.text(col, -row, symbol, ha='center', va='center', fontsize=12, fontweight='bold', color='black')
    ax.text(col, -row - 0.3, f"{atomic_number}", ha='center', va='center', fontsize=8, color='black')
    ax.text(col, -row + 0.3, f"{weight:.2f}", ha='center', va='center', fontsize=8, color='black')

# Save as PNG
plt.savefig("periodic_table.png", dpi=300, bbox_inches='tight')
plt.show()
```
